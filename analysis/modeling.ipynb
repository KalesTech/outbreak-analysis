## Target Definition

The target variable represents whether the next-day case count exceeds
the recent 7-day average, serving as a proxy for elevated outbreak
activity.

```
df["high_cases"] = (df["reported_cases"] > df["cases_ma7"]).astype(int)
```

## Train/Test Split

The dataset is split chronologically to preserve temporal structure.
Random shuffling is intentionally avoided.

```
split_index = int(len(df) * 0.8)

X = df[["cases_ma7", "symptom_ma7", "testing_ma7"]]
y = df["high_cases"]

X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]
```

## Feature Scaling

Standardization is applied to improve model stability and convergence.

```
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Model Training

A logistic regression classifier is trained as a simple, interpretable
baseline model.

```
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

## Model Evaluation

Model performance is evaluated using accuracy and a classification
report. Results should be interpreted cautiously given dataset size
and synthetic nature.

```
y_pred = model.predict(X_test_scaled)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## Interpretation

The model demonstrates modest predictive capability but is limited by
small sample size, synthetic data construction, and simplified features.
Performance is not intended to represent real-world outbreak prediction.

The primary goal of this model is methodological illustration rather
than operational accuracy.

## Limitations and Future Work

- Synthetic data limits external validity.
- More sophisticated temporal models (e.g., sequence models) could be explored.
- Additional data sources could improve feature richness.
- Model evaluation is constrained by dataset size.

Future work would focus on richer feature engineering and more robust
time-series modeling approaches.
