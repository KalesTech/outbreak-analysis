# Predictive Modeling

This notebook builds a simple predictive model using the synthetic outbreak
dataset. The goal is to demonstrate a complete modeling workflow, including
feature engineering, time-aware splitting, model training, evaluation, and
interpretation.

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/synthetic_outbreak.csv", parse_dates=["date"])
```

## Feature Engineering

This section constructs features suitable for time-series modeling.
All features are derived from past information only to avoid data leakage.

```python
df["cases_ma7"] = df["reported_cases"].rolling(7).mean()
df["symptom_ma7"] = df["symptom_rate"].rolling(7).mean()
df["testing_ma7"] = df["testing_volume"].rolling(7).mean()

df = df.dropna()
```

## Target Definition

The target variable indicates whether the current day's reported case count
exceeds the recent 7-day average, serving as a proxy for elevated outbreak
activity.

```python
df["high_cases"] = (df["reported_cases"] > df["cases_ma7"]).astype(int)
```

## Train/Test Split

The dataset is split chronologically to preserve temporal structure.
Random shuffling is intentionally avoided.

```python
split_index = int(len(df) * 0.8)

X = df[["cases_ma7", "symptom_ma7", "testing_ma7"]]
y = df["high_cases"]

X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]
```

## Feature Scaling

Standardization is applied to improve model stability and convergence.

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Model Training

A logistic regression classifier is trained as a simple, interpretable
baseline model.

```python
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

## Model Evaluation

Model performance is evaluated using accuracy and a classification report.
Results should be interpreted cautiously given dataset size and synthetic
data generation.

```python
y_pred = model.predict(X_test_scaled)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## Interpretation

The model demonstrates modest predictive capability but is constrained by
small sample size, synthetic data construction, and simplified features.
Performance is not intended to represent real-world outbreak prediction.

The primary purpose of this model is methodological illustration rather
than operational accuracy.

## Limitations and Future Work

- Synthetic data limits external validity.
- Additional temporal features and alternative models could be explored.
- Larger datasets would enable more robust evaluation.
- Results should not be interpreted as real-world outbreak forecasts.
